```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df.bought_insurance, test_size = 0.2, random_state = 20)
```

```python
len(X_train)
```

```python
# scale the age column , assume max age is 100, so divide age column by 100 so that all values will be 0 to 1
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age']/100
```

```python
X_train_scaled.head()
```

```python
# build neural network using tensorflow with 1 layer and input shape is 2, kernel initializer = 1 (initial weight start with 1), bias = 0
model = keras.Sequential([
    keras.layers.Dense(1, input_shape = (2,), activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer = 'zeros')
])

model.compile(
    optimizer = 'adam',
    loss ='binary_crossentropy',
    metrics = ['accuracy']
   )

model.fit(X_train_scaled, y_train, epochs = 5000)
```

```python
# Evaluate the model on test data
model.evaluate(X_test_scaled, y_test)
```

```python
# define sigmoid function 
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
```

```python
# define log loss function

def log_loss(y_true, y_predicted):
    eplison = 1e-15
    y_predicted_new = [max(i, eplison) for i in y_predicted]
    y_predicted_new = [min(i, 1-eplison) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new)+(1-y_true) * np.log(1-y_predicted_new))
```

```python
# define sigmoid function using numpy
import math
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12, 0,1]))
```

```python
# define gradient descent function

def gradient_descent(age, affordibility, y_true, epochs):
    
    w1 = w2 = 1 # assume initial weights start with 1
    bias  = 0 # assume initial bias start with zero
    rate = 0.1 # assume learning rate to be 0.5 
    n = len(age) # total no of rows equal to either age or affordibility
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        
        loss = log_loss(y_true,y_predicted) # loss is defined as per log loss or binary cross entropy 
        
        # code for derivative of w1 = w2  = 1/n (sum (age *(y_predicted - y_true)))
        # new w1 = w1-learning_rate*derivative of w. Transport age so that do product can be performed
        
        w1d = (1/n) * np.dot(np.transpose(age), (y_predicted-y_true))
        w2d = (1/n) * np.dot(np.transpose(affordibility), (y_predicted-y_true))
        
        bias_d = np.mean(y_predicted-y_true) # 1/n(sum(y_predicted-y_true))
        
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d
        
        print(f'Epoch:{i}, w1 : {w1}, w2 : {w2}, bias : {bias}, loss : {loss}')
              
    return w1, w2, bias
```

```python
# define gradient descent function

def gradient_descent(age, affordibility, y_true, epochs, loss_threshold):
    
    w1 = w2 = 1 # assume initial weights start with 1
    bias  = 0 # assume initial bias start with zero
    rate = 0.1 # assume learning rate to be 0.5 
    n = len(age) # total no of rows equal to either age or affordibility
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        
        loss = log_loss(y_true,y_predicted) # loss is defined as per log loss or binary cross entropy 
        
        # code for derivative of w1 = w2  = 1/n (sum (age *(y_predicted - y_true)))
        # new w1 = w1-learning_rate*derivative of w. Transport age so that do product can be performed
        
        w1d = (1/n) * np.dot(np.transpose(age), (y_predicted-y_true))
        w2d = (1/n) * np.dot(np.transpose(affordibility), (y_predicted-y_true))
        
        bias_d = np.mean(y_predicted-y_true) # 1/n(sum(y_predicted-y_true))
        
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d
        
        print(f'Epoch:{i}, w1 : {w1}, w2 : {w2}, bias : {bias}, loss : {loss}')
        
        if loss<=loss_threshold:
            break
              
    return w1, w2, bias
```
